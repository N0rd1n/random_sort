```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def check_sort(a):
    for i in range(1, len(a)):
        if a[i-1] > a[i]:
            return False
    return True

def shit_sort(a):
    list_sorted = False
    step = 0
    while not list_sorted:
        step += 1
        a = random.sample(a, len(a))
#         print("Step {step}: {a}".format(step = step, a = a))
        list_sorted = check_sort(a)
    return a, step
```

```python
shit_sort(range(4))
```

```python
result = list()
for i in range(100):
    a, steps = shit_sort(range(10))
    result.append(steps)
series_result = pd.Series(result)

series_result.plot.hist(grid=True, bins=20, rwidth=0.9,
                   color='#607c8e')
plt.title('Shit Sort attempts')
plt.xlabel('Counts')
plt.ylabel('Commute Time')
plt.grid(axis='y', alpha=0.75)
```
